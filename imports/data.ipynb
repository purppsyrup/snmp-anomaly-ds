```python
import pandas as pd

data = pd.read_csv("all_data.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
```

```python
grouped_data = data.groupby('class').agg(['mean', 'median', 'std'])
```

```python
summary_stats = grouped_data.describe()
```

```python
print(summary_stats)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
data_encoded = data.copy()
data_encoded['class'] = label_encoder.fit_transform(data['class'])

# Select only numeric columns for correlation calculation
data_numeric = data_encoded.select_dtypes(include=['float64', 'int64'])

# Calculate correlation matrix
corr_matrix = data_numeric.corr()

# Plot heatmap using seaborn
plt.figure(figsize=(20, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```
